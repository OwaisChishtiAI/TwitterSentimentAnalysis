## Let's save these tweets in file in our directory, for sentiment analysis.

```python
# Importing Libraries
from tweepy import Stream
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import time

# Navigate to your twitter apps keys and use them.
ckey="aJ8N5zsMXnaRBJpJUBPfsVN2m"
csecret="mMBLEAw2YiaPST9EEXsLbKBw68LfzWQU5yiVB8FkTZn9a6B27c"
atoken="587454349-46cFBwggREZMn9CFOlAy0t9Arh68hWUDKbTuA44W"
asecret="BP8dC80EDGpJvackYNs6Qu6BJ9rShTLwHTB09z6XCZiIP"

class listener(StreamListener):

    def on_data(self, data):
        try:
            tweet = data.split(',"text":"')[1].split('","source')[0]
            print (tweet)
            tweet = str(tweet)
            with open("tweets.txt", "a") as f:
                f.write(tweet)
                f.write("\n")
            return(True)
        except BaseException as e:
            print ("Network Error", str(e))
            time.sleep(2)
    def on_error(self, status):
        print (status)

auth = OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)

twitterStream = Stream(auth, listener())
twitterStream.filter(track=["car"])
```

## Finally, we are about to perform sentiment analysis with 'textblob'.

** TextBlob is a Python (2 and 3) library for processing textual data. It provides a simple API for diving into common natural language processing (NLP) tasks such as part-of-speech tagging, noun phrase extraction, sentiment analysis, classification, translation, and more. **

** download textblob as follows,**
* open command prompt in windows or teminal in Linux/MAC.
* .>>> pip install -U textblob
* .>>> python -m textblob.download_corpora 

last line written above will download 'Bag of Words', to determine the relevant and irrelevant words.

Use link to read more about:
1) [Bag of Words](https://en.wikipedia.org/wiki/Bag-of-words_model)
2) [textblob Documentation](http://textblob.readthedocs.io/en/dev/)

```python
from textblob import TextBlob
i = 0

def read_data():
    tweets = open("tweets.txt", "r").readlines()
    return tweets
def Sentiment_analysis(tweet):
    analysis = TextBlob(tweet)
    return (analysis.sentiment)

tweets = read_data()


for tweet in tweets:
    print(tweet)
    analysis = Sentiment_analysis(tweet)
    print(analysis)
    print("\n"+"------")
    i += 1
    if i == 2:
        break
```

** Sentiment(polarity=x.xx, subjectivity=x.xx), in this line 'polarity' shows how good or bad the sentence is, and 'subjectivity' shows that the reality of sentence as per universe. **

** As, you can that first tweet seems to be good thus (polarity = 0.275) and second tweet seems to be bad that's why (polarity = -0.33) **
